```python
def plant_diag():

    diagnoses = {
        'yellow leaves': {
            'problem':'Overwatering',
            'treatment': 'Reduce watering, ensure pot has drainage holes, and let soil dry before watering again.'},
        'drooping leaves': {
            'problem': 'Underwatering',
            'treatment': 'Water the plant thoroughly and set a regular watering schedule.'},
        'brown tips': {
            'problem': 'Low humidity or over-fertilizing',
            'treatment': 'Increase humidity (mist leaves) and reduce fertilizer use.'},
        'wilting': {
            'problem': 'Root stress or improper watering',
            'treatment': 'Check soil moisture and adjust watering. Avoid overwatering.'},
        'slow growth': {
            'problem': 'Lack of nutrients or low light',
            'treatment': 'Move plant to brighter area and apply balanced fertilizer.'},
        'white spots': {
            'problem': 'Pests (fungus or insects)',
            'treatment': 'Clean leaves and use mild insecticidal soap or neem oil.'},
        'leaf curling': {
            'problem': 'Heat stress or underwatering',
            'treatment': 'Move plant away from heat source and adjust watering.'},
        'mold on soil': {
            'problem': 'Overwatering and poor airflow',
            'treatment': 'Let soil dry, improve ventilation, and remove top soil layer.'}
    }
    
    print('Select on of the below symptoms: ')
    print('1. Yellow Leaves')
    print('2. Drooping Leaves')
    print('3. Brown Tips')
    print('4. Wilting')
    print('5. Slow Growth')
    print('6. White Spots')
    print('7. Leaf Curling')
    print('8. Mold On Soil')
    user_input=input('\nEnter symptom number: ')

    if user_input == '1':
        diag = diagnoses['yellow leaves']
    elif user_input == '2':
        diag = diagnoses['drooping leaves']
    elif user_input == '3':
        diag = diagnoses['brown tips']
    elif user_input == '4':
        diag = diagnoses['wilting']
    elif user_input == '5':
        diag = diagnoses['slow growth']
    elif user_input == '6':
        diag = diagnoses['white spots']
    elif user_input == '7':
        diag = diagnoses['leaf curling']
    elif user_input == '8':
        diag = diagnoses['mold on soil']
    else:
        print('The entered symptom number is incorrect!')
        return

    print(f'Problem: {diag['problem']}')
    print(f'Treatment: {diag['treatment']}')
```

```python
plant_diag()
```
