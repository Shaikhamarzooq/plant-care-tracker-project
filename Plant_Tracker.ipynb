```python
import csv
from datetime import datetime

def activity_capture():

    """
    This function will allow selecting the plant by plant ID and then select the desired activity
    """

#Below will read the plant_id and plant_name from the plant_data file

    plants = []
    with open('../plant-care-tracker-project/plants_data.csv','r',newline = '') as file:
        content = csv.DictReader(file)
        print(f'\nPlant List:\n')
        for i in content:
            plants.append(i)
            print(f'{i['plant_id']} | {i['plant_name'].upper()}')
          
#Below will read user's input, where he/she will select a plant and the activity to be performed
    
    plant_id = input('\n**Enter plant ID from the above list: ')
    
    plant_name = ''
    
    for i in plants:
        if plant_id == i['plant_id']:
            plant_name = i['plant_name'].upper()
            break

#Below will check if the plant ID is valid
    
    if plant_name:
        print(f'\n**The selected plant is {plant_id} : {plant_name}')
    else:
        print('The entered plant ID is incorrect!')
        return
        
    activity = input('\n1.Watering. \n2.Fertilizing. \n3.Repotting. \n4.Pruning \n\n**Enter the number/s from the above activity separated by comma : ').split(",")
    
    plant_activity = {'date': datetime.now().strftime('%d-%m-%Y'),
                      'plant_id': plant_id,
                      'watered': False,
                      'fertilized': False,
                      'repotted': False,
                      'pruned': False}

#Below will check if the activity is valid

    for i in activity:      
        if i == '1':
            plant_activity['watered']= True
        elif i == '2':
            plant_activity['fertilized']= True
        elif i == '3':
            plant_activity['repotted']= True
        elif i == '4':
            plant_activity['pruned']= True  
        else:
            return f'Activity enter is incorrect!'

#Below will append the activity into the activity_log file   
    
    activity_log_headers = ['date','plant_id','watered','fertilized','repotted','pruned']  
    with open('../plant-care-tracker-project/activity_log.csv', 'a', newline='\n') as activity_log:
        plantlog_writer = csv.DictWriter(activity_log, fieldnames = activity_log_headers)
        plantlog_writer.writerow(plant_activity)

    print(f'\n Activity log been saved')
    print(f'\n {plant_activity}')


def main():
    print('1. Add a new plant to the collection.')
    print('2. Record a plant care activity.')
    print('3. View plants due for care.')
    print('4. Search plants by name or location.')
    print('5. View all plants.')

    user_input = input(f'\nChoose one of the above actions: ')

    if user_input == '1':
        add_plant()
    elif user_input == '2':
        record_plant()
    elif user_input == '3':
        activity_capture()
    elif user_input == '4':
        search_plant()
    elif user_input == '5':
        view_all_plants()
    else:
        return f'The entered option number is incorrect!'
```

```python
main()
```
